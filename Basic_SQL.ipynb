
## ANY/IN/EXISTS
**Se la subquery ritorna piu' di un singolo valore**

*si deve utilizzare un operatore in grado di supportare il confronto con set*

- Operatori quantificati: `<attr><op-rel>[ANY|ALL]<subquery`
- Operatore di set: `<attr> [NOT] IN <subquery>`
- Quantificatore esistenziale : `[NOT] EXISTS <subquery>`

````
# TIPO [ANY/ALL]
# Per esprimere una condizione che e' vera quando 
# qualsiasi (ANY) opppure tutti (ALL) gli elementi
# restituiti dalla subquery fanno verificare la condizione
# i.e., condizione risulta TRUE
 
# ANY
# Nome degli studenti che hanno sostenuto l’esame del corso C1

q = '''
SELECT SNome
FROM S join E ON S.Matr = E.Matr
WHERE CC='C1'
'''
con.execute(q) 
con.fetchdf()
````

````
q = '''
SELECT SNome
FROM S natural JOIN E
WHERE E.CC = 'C1';
'''
con.execute(q) 
con.fetchdf()
````

````
# ALL
# Studenti con anno di corso piu` basso
q = '''
SELECT *
FROM S
WHERE ACorso <= ALL ( SELECT ACorso
          					  FROM S);
'''
con.execute(q) 
con.fetchdf()
````

````
# versione con MIN
q = '''
SELECT *
FROM S
WHERE ACorso <=  ( SELECT MIN(ACorso)
          				    FROM S);
'''
con.execute(q) 
con.fetchdf()
````

### EXISTS

````
q = '''
SELECT *
FROM S
'''
con.execute(q) 
con.fetchdf()			
````

````
# QUALIFICATORE ESISTENZIALE:

# EXISTS

# Mettendo in relazione subquery e query estenza con una clausola
# che è vera solo se per la tupla corrente il risultato della subquery
# restituisce qualcosa (i.e., non è nullo).
# Da qui: "EXISTS", cioè ESISTE qualcosa che rende TRUE quel predicato.

# Ad esempio:
# Nome degli studenti che hanno sostenuto l’esame del corso C1:

q = '''
SELECT SNome
FROM S
WHERE EXISTS (SELECT *
              FROM E
              WHERE E.Matr=S.Matr
              AND E.CC = 'C1')
'''
con.execute(q) 
con.fetchdf()			  
````

````
# Allo stesso modo, il negato della clausola EXISTS:

# NOT EXISTS (<subquery>)

# ha valore true se e solo se l’insieme di valori restituiti da <subquery> e' vuoto.

# Ad esempio:
# Nome degli studenti che non hanno sostenuto l’esame del corso C1
q = '''
SELECT SNome
FROM S
WHERE NOT EXISTS ( SELECT *
                   FROM E
                   WHERE E.Matr=S.Matr 
                   AND E.CC='C1' )
'''
con.execute(q) 
con.fetchdf()

# Tip:
# Se state usando EXISTS senza una subquery correlata,
# molto probabilmente state commettendo un errore:

# Perche'?
# (hint: cosa restituisce la subquery?) 
````

### Riduzione di query innestate

Le query innestate formulate con i seguenti operatori 
si possono ridurre a query di join equivalenti 

(stessa risposta per ogni possibile istanza della base di dati):

- IN
- ANY (con qualsiasi operatore di confronto) 
- EXISTS con subquery correlata

````
# Nome degli studenti che hanno sostenuto l’esame del corso C1

# 1. 
q = '''
SELECT SNome 
FROM S
WHERE Matr IN ( SELECT Matr
                FROM E
                WHERE CC='C1')
'''
con.execute(q) 
con.fetchdf()
````

````
# 2. 
q = '''
SELECT SNome 
FROM S
WHERE Matr = ANY (SELECT Matr
                  FROM E
                  WHERE CC='C1')
'''
con.execute(q) 
con.fetchdf()
````

````
# 3.
q = '''
SELECT SNome 
FROM S
WHERE EXISTS ( SELECT *
                FROM E
                WHERE E.Matr=S.Matr
                AND CC='C1')
'''
con.execute(q) 
con.fetchdf()
````

````
# QUERY EQUIVALENTE CON JOIN
q = '''
SELECT SNome
FROM E,S
WHERE E.Matr=S.Matr 
AND E.CC='C1'
'''
con.execute(q) 
con.fetchdf()
````

INVECE:
 
Le query innestate formulate con i seguenti operatori 
non si possono ridurre:

- NOT IN
- ALL (con qualsiasi operatore di confronto) 
- NOT EXISTS con subquery correlata

````
# Ad esempio:
# Nome degli studenti che non hanno sostenuto l’esame del corso C1

# 1. 
q = '''
SELECT SNome 
FROM S
WHERE Matr NOT IN ( SELECT Matr
                    FROM E
                    WHERE CC='C1')
'''
con.execute(q) 
con.fetchdf()
````

````
# 3.
q = '''
SELECT SNome 
FROM S
WHERE NOT EXISTS (SELECT *
                  FROM E
                  WHERE E.Matr=S.Matr
                  AND CC='C1')
'''
con.execute(q) 
con.fetchdf()

# NON SI POSSONO ESPRIMRE CON JOIN
````

## GROUP BY

In una istruzione SELECT e' possibile formare 
dei gruppi di tuple che hanno lo stesso valore 
di specificati attributi, tramite la clausola
- `GROUP BY`


SINTASSI:
```
SELECT [DISTINCT|ALL] <lista-SELECT> FROM <lista-FROM>
[WHERE <condizione>]
[GROUP BY <lista-group> ]
```


**IMPORTANTE**

Il risultato della SELECT e' un unico record per ciascun gruppo

````
# La query di prima:
q='''
SELECT Matr, AVG(Voto)
FROM E
GROUP BY Matr
'''
con.execute(q).fetchdf()
````

````
# La query di prima:
q='''
SELECT *
FROM E
'''
con.execute(q).fetchdf()
````

````
# “Voto massimo e minimo ottenuto per ogni studente, escludendo il corso C1”
q='''
SELECT Matr, MAX(Voto), MIN(Voto) 
FROM E
WHERE CC <> 'C1'
GROUP BY Matr
'''
con.execute(q).fetchdf()
````

````
# Codice e nome di un corso, e relativo numero di esami sostenuti
q='''
SELECT C.CC, C.CNome, COUNT(*) as num_esami
FROM E,C
WHERE E.CC=C.CC
GROUP BY C.CC, C.CNome
'''
con.execute(q).fetchdf()
````

## Clausola HAVING

In presenza di una clausola WHERE, il raggruppamento riporta solo i valori che soddisfano la condizione della WHERE

````
# Numero esami e somma dei voti per ogni studente, escludendo il corso C2
q='''
SELECT Matr, COUNT(*), SUM(Voto)
FROM E
WHERE CC <> 'C2'
GROUP BY Matr
'''
con.execute(q).fetchdf()
# ma non restituisce alcuna informazione sulla 
# matricola M6, in quanto nessuna tupla di E
# corrispondente a M6 soddisfa la condizione
````

````
# Dobbiamo usare GROUP BY ALL per considerare anche i NULL
q='''
SELECT Matr,COUNT(*),SUM(Voto) 
FROM E
WHERE CC <> 'C2'
GROUP BY ALL Matr
'''
# con.execute(q).fetchdf()
````

````
# Se vogliamo un risultato con lo 0 al posto del NULL
q='''
SELECT Matr, COUNT(*), ISNULL(SUM(Voto),0) 
FROM E
WHERE CC <> 'C2'
GROUP BY ALL Matr
'''
# con.execute(q).fetchdf()
````

Se volessimo gli studenti con una media maggiore di 28?
  
**NON POSSIAMO FARE QUANTO SEGUE:**

````
q='''
SELECT Matr,AVG(VOTO) 
FROM E
WHERE VOTO > 28
GROUP BY Matr
'''
con.execute(q).fetchdf()
````

### HAVING
La clausola HAVING e' l’equivalente dellaclausola WHERE
applicata a gruppi di tuple: 

ogni gruppo costruito tramite GROUP BY 
fa parte del risultato dell’interrogazione 
se e solo se 
il predicato specificato nella clausola HAVING risulta soddisfatto.


**SINTASSI:**
```
[GROUP BY <lista-group> ]
[HAVING <condizione>]
```

````
# Selezionare studente e media, degli studenti con media maggiore di 28
q='''
SELECT Matr, AVG(VOTO) 
FROM E
GROUP BY Matr
HAVING AVG(VOTO) > 28
'''
con.execute(q).fetchdf()
````

````
# Media dei voti per ogni esame sostenuto da piu' di due studenti
q='''
SELECT CC, AVG(Voto)
FROM E
GROUP BY CC
HAVING COUNT(*) > 2
'''
con.execute(q).fetchdf()
````

````
# Corsi per cui ci sono almeno due studenti con lo stesso voto
q='''
SELECT CC, Voto
FROM E
GROUP BY CC, Voto
HAVING COUNT(Matr) >= 2
'''
con.execute(q).fetchdf()
````

````
q='''
SELECT *
FROM E
'''
con.execute(q).fetchdf()
````

````
q='''
SELECT E.Matr,COUNT(*) as num, AVG(Voto)
FROM E join S on E.Matr=S.Matr
WHERE CC <> 'C2'
GROUP BY E.Matr
ORDER BY AVG(Voto) DESC
'''
con.execute(q).fetchdf()
````

````
# Provare aggiungendo S.SNome alla GROUP BY

q='''
SELECT E.Matr, COUNT(*) as num, AVG(Voto)
FROM E join S on E.Matr=S.Matr
WHERE CC <> 'C2'
GROUP BY E.Matr, S.SNome
ORDER BY E.Matr
'''
con.execute(q).fetchdf()
````

````
# Rimuovere di nuovo S.SNome alla GROUP BY
# Aggiungere SELECT anche S.SNome

q='''
SELECT E.Matr, S.SNome, COUNT(*) as num, AVG(Voto)
FROM E join S on E.Matr=S.Matr
WHERE CC <> 'C2'
GROUP BY E.Matr, S.SNome
ORDER BY E.Matr
'''

con.execute(q).fetchdf()
````
